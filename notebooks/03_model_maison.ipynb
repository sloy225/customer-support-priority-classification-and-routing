```python
import os
import sys

# 1. Détection dynamique de la racine du projet
dossier_actuel = os.getcwd()
racine_projet = None

# On remonte les dossiers jusqu'à trouver celui qui contient le dossier 'models'
while dossier_actuel != os.path.dirname(dossier_actuel):
    if os.path.exists(os.path.join(dossier_actuel, "models")):
        racine_projet = dossier_actuel
        break
    dossier_actuel = os.path.dirname(dossier_actuel)

# Fallback : Si la détection dynamique échoue, on force ton chemin absolu Windows
if not racine_projet:
    racine_projet = r"C:\Users\yamba\Documents\projetDL\customer-support-priority-classification-and-routing"

# 2. On insère le chemin à l'index 0 (priorité absolue pour Python)
if racine_projet not in sys.path:
    sys.path.insert(0, racine_projet)

print("🎯 Racine officielle du projet verrouillée :", racine_projet)
```
